```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/Task2hp/HPSpells.csv")
```

```python
data
```

```python
data.head()
```

```python
data=data.drop(data[data.Type=='Hex'].index)
```

```python
data
```

```python
data=data.dropna(subset=["Light"])
```

```python
data
```

```python
freq={}
for Type in data.Type.unique():
  c=0
  for item in data.Type :
    if item==Type:
      c+=1
    freq[Type]=c
freq

plt.figure(figsize=(20,7))

plt.bar(range(len(data.Type.unique())),list(freq.values()))
plt.xticks(range(len(freq)),data.Type.unique())
plt.show()
```
